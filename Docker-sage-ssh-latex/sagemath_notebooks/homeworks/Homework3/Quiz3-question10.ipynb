$ \text{det}(A - \lambda I) = 0 $

Let $A - \lambda I$ be the matrix:

$ A - \lambda I = \begin{bmatrix} 2 - \lambda & 1 \\ 1 & 2 - \lambda \end{bmatrix} $

Now, find the determinant of this matrix and set it equal to zero:

$ \text{det}(A - \lambda I) = (2 - \lambda)(2 - \lambda) - (1)(1) = (2 - \lambda)^2 - 1 $

Now, set the determinant equal to zero:

$ (2 - \lambda)^2 - 1 = 0 $

$ (2 - \lambda)^2 = 1 $

Taking the square root of both sides:

$ 2 - \lambda = \pm 1 $

1. When $2 - \lambda = 1 $:

2. When $2 - \lambda = -1 $:
   $ \lambda_2 = 2 + 1 = 3 $

So, the eigenvalues of matrix $A$ are $\lambda_1 = 1$ and $\lambda_2 = 3$.
